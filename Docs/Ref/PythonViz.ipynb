```python
# @widgets.interact  #Real-time update
@widgets.interact_manual   #Manual update
def projectile(v0=50,v1=10, option=True, string='20'):
    t = np.linspace(0,10)
    g = 9.81
    y = v0*t - 0.5*g*t**2
    y1 = v1*t - 0.5*g*t**2
    plt.plot(t,y,"-",label='v0 = {} $m/s$'.format(v0))
    plt.plot(t,y1,".",label='v1 = {} m/s'.format(string))
    plt.ylim(0,)
    plt.xlabel("t [s]")
    if option:
        plt.ylabel("height [m]")
    plt.legend()
    plt.show()
```
