````python
#importing some necessary packages
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
````

````python
#Loading the dataset
data = pd.read_csv('../../Lab1/Romil/Data/pima.csv',header =None)
data.head()
````

````python
#knowing the shape of data
data.shape
````

````python
#Extracting X and y from dataset
X = data.drop(8,axis=1)
y = data[8]
X.shape,y.shape
````

````python
#Splitting dataset into train and test 
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
````

````python
#Initializing model as sequential
model = tf.keras.models.Sequential()
````

** Adding layers**

````python
model.add(tf.keras.layers.Dense(12,input_dim = 8, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(8, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation = tf.nn.sigmoid))
````

** configuration for training, adding optimizer, loss and metrics**

````python
model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
````

** Saving checkpoint and model when accuracy is increased **


````python
from keras.callbacks import ModelCheckpoint
ckpt_model = 'pima-weights.best.hdf5'
checkpoint = ModelCheckpoint(ckpt_model, 
                            monitor='val_acc',
                            verbose=1,
                            save_best_only=True,
                            mode='max')
callbacks_list = [checkpoint]
````

````python
# Fitting the model
history = model.fit(x_train,
                    y_train,
                    validation_data=(x_test, y_test),
                    epochs=500,
                    batch_size=10,
                    callbacks=callbacks_list,
                    verbose=0)
````

* Optimizer = adam
 1. activation = relu,accuracy = 0.8580,validation acc = 0.6889
 2. activation = sigmoid,accuracy = 0.8113,validation acc = 0.7322
 3. activation = tanh,accuracy = 0.7490,validation acc = 0.6732




### Model accuracy

````python
# Model accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
````

```
In above plot, we have plotted the accuracy per epoch for train and testing dataset.
As you can see, there are no much more accuracy difference between train and test which is good
```

### Model Loss

````python
# Model Losss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
````

```
As we can see that model loss has decrease to min level at about 40 epoch. after that it stops to decrease.
```

````python
# print final accuracy
scores = model.evaluate(x_test, y_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
````
