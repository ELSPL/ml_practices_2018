```python
# importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.mixture import GaussianMixture
```

```python
# loading the data
mush = pd.read_csv('../../Lab1/Heenal/Data/mushroom_new.csv')
```

```python
mush.head()
```

```python
X = mush[['gill-color', 'cap-color']]
y = mush['class']
```

```python
X.shape, y.shape
```

```python
X = mush.drop(['class'], axis=1)
sns.pairplot(X)
```

```python
sns.heatmap(mush.corr(), annot=True)
```

```python
gmm = GaussianMixture(n_components=2)
```

```python
gmm.fit(X)
```

```python
labels = gmm.predict(X)
```

```python
X['labels'] = labels
x0 = X[X['labels'] == 0]
x1 = X[X['labels'] == 1]
```

```python
x0.shape
```

```python
plt.scatter(x0['gill-color'], x0['cap-color'], c='r')
plt.scatter(x1['gill-color'], x1['cap-color'], c='b')
```
