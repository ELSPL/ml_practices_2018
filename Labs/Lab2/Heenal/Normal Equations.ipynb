```python
from sklearn import preprocessing
import numpy as np
from sklearn import datasets
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston = datasets.load_boston()
```

```python
boston.feature_names
```

## Univariate

```python
X,Y = boston.data[:,6:7],boston.target
```

```python
X.shape
```

```python
# appending a column of ones in the feature matrix for theta0
X_new = np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
Y.shape
```

```python
X_T = np.transpose(X_new)
X_T.shape
```

```python
temp = np.matmul(X_T, X_new)
temp_inv = np.linalg.inv(temp)
theta = np.matmul(np.matmul(temp_inv, X_T), Y)
```

```python
theta
```

```python
slope = theta[1]
c = theta[0]

fig, ax = plt.subplots()
ax.scatter(X, Y)
x_min, x_max = ax.get_xlim()
y_min, y_max = c, c + slope*(x_max-x_min)
ax.plot([x_min, x_max], [y_min, y_max], color='r')
ax.set_xlim([x_min, x_max])
```
