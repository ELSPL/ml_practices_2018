# Statistical learning
### Calculating the coefficients for the equation of line



```python
m = (((X*Y).mean()) - (X.mean()*Y.mean())) / (((X*X).mean()) - (X.mean()*X.mean()))
```

```python
c = Y.mean() - m * X.mean()
```

```python
m
```

```python
c
```

```python
fig, ax = plt.subplots()
ax.scatter(X, Y)

# need a slope and c to fix the position of line
slope = m


x_min, x_max = ax.get_xlim()
y_min, y_max = c, c + slope*(x_max-x_min)
ax.plot([x_min, x_max], [y_min, y_max], color='r')
ax.set_xlim([x_min, x_max])
```
