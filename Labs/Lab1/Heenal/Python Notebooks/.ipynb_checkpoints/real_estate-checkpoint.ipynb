````python
# Importing all necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
estate = pd.read_excel('../Data/Real estate valuation data set.xlsx')
````

````python
estate
````

````python
estate.describe()
````

##  Figuring out number of null values in each column

````python
estate.isnull().sum()
````

** Hence, we observe that no column has any null values. **

# Plotting the correlation matrix using a heatmap 

````python
sns.set(rc={'figure.figsize':(15,8)})
ax = sns.heatmap(estate.corr(),annot=True)
````

** Conclusion : **
 ```
1. The housing prices depend on features such as distance to nearest MRT station, number of convenience stores, latitude, longitude and the house age.
2. The features such as distance to the nearest MRT station and longitude have a trong negative correlation.
 ```

### Converting transaction date to only year format for year-wise analysis 

````python
estate.dtypes
````

````python
est = estate
est['X1 transaction date'] = est['X1 transaction date'].astype(int)
est
````

# Comparing the effect of number of convenience stores on the housing prices in different years

````python
g = sns.FacetGrid(est, col="X1 transaction date")
g.map(plt.scatter, "X4 number of convenience stores", "Y house price of unit area")
g.add_legend();
````

````python
sns.lmplot(x="X4 number of convenience stores", y="Y house price of unit area", hue="X1 transaction date", data=est, markers=["o", "x"]);
````

### Observation :
    The prices rose by a small margin as the year progressed.

# Comparing the effect of distance to the nearest MRT station on the housing prices in different years

````python
sns.lmplot(x="X3 distance to the nearest MRT station", y="Y house price of unit area", hue="X1 transaction date", data=est, markers=["o", "x"]);
````

### Observation :
    The distance does act as a significant factor affecting prices of houses over years.

# Effect of the distance to nearest MRT station on the housing prices

````python
g = sns.jointplot("X3 distance to the nearest MRT station", "Y house price of unit area", data=estate, space=0, color="r", kind='reg')
````

### Observation:
    Then price of house decreases as the distance to nearest MRT station increases.

# Effect of the age of house on the price of house

````python
g = sns.jointplot("X2 house age", "Y house price of unit area", data=estate, space=0, color="r", kind='reg')
````

### Observation:
    The age of the house does not affect the price of house within the age range of 0-45.
