````python
# ID-wise sorted data

absent.sort_values('ID')
````

# Analysis of season affecting the absenteeism count
```
Season = 1 , Summer
         2 , Autumn
         3 , Winter
         4 , Spring
```

````python
# Season-wise count of absents per ID (Assuming that Absenteeism time in hours = 0 refers to Not Absent)

absent[absent['Absenteeism time in hours'] != 0].groupby(['ID']).Seasons.value_counts().unstack()
````

````python
# Sum total of absents in all seasons

absent[absent['Absenteeism time in hours'] != 0].groupby(['ID']).Seasons.value_counts().unstack().median()
````

````python
# Plotting the above count

absent[absent['Absenteeism time in hours'] != 0].groupby(['ID']).Seasons.value_counts().unstack().plot(kind='bar')
````

### Observation :

Season 2 i.e. Autumn observes the maximum average absences as compared to other seasons.

# Relation of drinking habits with absenteeism count

``` 
Social drinker = 0 , does not drink
                 1 , drinks
                 
```

````python
absent.groupby(['Social drinker'])['Absenteeism time in hours'].value_counts().sort_index()
````

````python
absent.groupby(['Social drinker'])['Absenteeism time in hours'].value_counts().sort_index().plot(kind='bar')
````

### Observation:
The employees' drinking habits do not affect their absenteeism count at large because the time of absence in hours for both drinkers and non-drinkers falls in the range of 0-16 hours approximately.
