# Information about dataset
```
Target Class:
    1 = Kama
    2 = Rosa
    3 = Canadian
To construct the data, seven geometric parameters of wheat kernels were measured: 
1. area A, 
2. perimeter P, 
3. compactness C = 4*pi*A/P^2, 
4. length of kernel, 
5. width of kernel, 
6. asymmetry coefficient 
7. length of kernel groove. 
```  

````python
#importing necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import scipy.stats as st
````

````python
#reading csv file
names = ['Area','Perimeter','Compactness','length_kernel','width_kernel','asy_coefficient','len_kernel_groove','target']
data = pd.read_csv('../Data/seeds_dataset.csv',header=None,names=names)
data.head()
````

** Target variable contains same instances of all the value in target attributes. That is why dataset is not biased. It would be good for classification.**

## Correlational map

````python
sns.heatmap(data.corr(), annot= True)
````

### Findings:
```
1. As we know the area is more related to length, width and perimeter. Above map exactly shows us that.
2. Also, length of kernel groove is related to seed's area, perimeter and kernel length.
```

## Analysing seed's dimensional features

````python
sns.set(rc={'figure.figsize':(15,8)})
plt.subplot(2,2,1)
x = sns.boxplot('target','length_kernel',data = data)
plt.subplot(2,2,2)
x = sns.boxplot('target','width_kernel',data = data)
plt.subplot(2,2,3)
x = sns.boxplot('target','Area',data = data)
plt.subplot(2,2,4)
x = sns.boxplot('target','Perimeter',data = data)
````

## Observation:
```
From all the above graph, it is very clear that Canadian wheat seeds has minimum ranges of length, width, Area, Perimeter while Rosa wheat seeds has maximum of it.
```

````python
x = sns.jointplot('target','len_kernel_groove', data = data,kind = 'kde')
````

## Analysing compactness
